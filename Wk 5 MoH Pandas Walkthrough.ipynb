# Pandas Demo

This demo works through importing pandas and using it to research a csv file.  CSV file must be found and imported.

```python
# Review class recording to find CSV on kaggle.  Most open CSV in VSC prior to importing.  Is there a military awards csv, MoH?

# Iports pandas and changes the name to shorthand pd
import pandas as pd

#create a dataframe object from a csv
# Takes already sorted CSV data and resorts for our use.
# tell pandas to read a file name within the () and 
moh_df = pd.read_csv('medal_of_honor.csv', delimiter=",")

moh_df
```

```python
# Prints top 10 rows plus header
moh_df.head(10)
```

```python
# Prints the min, mean, and max of a time, date, year...
print(moh_df['awarded.date.year'].min())
print(moh_df['awarded.date.year'].mean())
print(moh_df['awarded.date.year'].max())
print(moh_df['birth.date.year'].min())
print(moh_df['birth.date.year'].mean())
print(moh_df['birth.date.year'].max())
# The -1 printed reflects unavailable data in these columns.  Potetially becasue data was unavailable, especially for birth year.
# The decimal numbers reflect the math done to achive a mean of the numbers, not an actual mean year.
```

```python
# Gives arandom sample of data within the provided range
moh_df.sample(10)
```

```python
# I think this breaks out various aspects of data, like proportions
moh_df.describe(include='all')
```

```python
# Prints the count of a particualr vaule within a column
moh_df['military record.organization'].value_counts()
```

```python
# Prints the top X counts of a particular value.  In this case the 10 names that occur the most.
moh_df['name'].value_counts()[:10]
```

```python
# Visualization.  15 vaule counts, plots out as a bar horizontl, gives title of plot.  
# Pandas is great at making good looking visualizations.  Refer to pandas website.
# This cell created a bar graph numbers of MoH by branch.
moh_df['military record.organization'].value_counts()[:15].plot(kind="barh", title="Most MoH by Branch")
```

```python
moh_df['military record.organization'].value_counts()[:15].plot(kind="pie", title="Most MoH by Branch")
```

```python
moh_df['military record.rank'].value_counts()[:15].plot(kind="barh", title="Most MoH by Rank")
```

```python
moh_df['military record.rank'].value_counts()[:15].plot(kind="pie", title="Most MoH by Rank")
```
