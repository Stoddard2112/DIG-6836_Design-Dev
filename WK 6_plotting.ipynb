```python
V = {1, 2, 3, 4, 5}
E = {(1, 2), (1, 4), (2, 5), (3, 4), (4, 5)}

import networkx as nx

G = nx.Graph()
G.add_nodes_from(V)
G.add_edges_from(E)

import matplotlib.pyplot as plt

nx.draw_networkx(G, font_color="white")
plt.axis('off');
```
