```python
# This indexes the imported csv, orders it, by a chosen catagory.  In this case the year of death.  Good for analysis claims over time.  Year then alphabetical name
df = pd.read_csv('amac_inter_data_clean.csv')
df = df.set_index('Year of Death')
df.head()
```

```python
# WIth year as index, we can slice out particualr years to sirt data
df.loc[1919].head()
```

```python
# sub sets, years and month 
df.loc[1918, 'Month of Death'].head()
```

```python
# Logically letting us take catagories.  Here, break down by State, choosing Florida.  Seperate datasets within overall data set.
state = df.groupby('field_place_of_entry')
state.get_group('Florida').head()
```

```python
# Comparative analysis of groups, here user rating of fction vs nonfiction, could also do price
state['Branch'].mean()
```

```python
# running genre comp using groupby of a comlex grouping.  Grouped by year and genre.  Able to count number of unique titles in each group.
amac_state_branch_comparison = df.groupby(['field_place_of_entry','Branch'])['Last Name'].count()
amac_state_branch_comparison.head()
```

```python
print(amac_state_branch_comparison)
```

```python
# This plots the gnere groupby data over years
amac_state_branch_comparison.groupby(['Branch']).plot(legend=True)
```
