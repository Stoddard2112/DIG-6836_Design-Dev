# Stage Three: Clean Your Data

In an attempt to clean up the data, I'd like to remove the -1's from the sheet and replace it with NaN.  This should make the print out of the minimum year awarded cleaer.  

The -1 value was cleaned, but a new issue was found.  The creator of the data did not use a 4 digit year for each entry.  The latest print returned "63" as the earliest year awarded.  Since the MoH was not awarded in 63 CE, this is likely shorthand for 1863 or 1963.  (Upon further resarch into the data, this was 1863 for the awarding of Pvt. Platt during the Civil War.)


```python
moh_df = pd.read_csv('medal_of_honor.csv', na_values = '-1')

print(moh_df['awarded.date.year'].min())
print(moh_df['awarded.date.year'].max())
```

# Stage Four: Plot Your Data

The following three plots visualize different data in different ways.  Plot 1 is a horizontal bar graph of MoH awarded by branch of service.  Plot 2 is a pie chart of MoH awarded by rank.  Plot 3 is an area plot of MoH awarded by year.

```python
# Plot 1
moh_df['military record.organization'].value_counts()[:5].plot(kind="barh", title="Most MoH by Branch")
```

```python
# Plot 2
moh_df['military record.rank'].value_counts()[:15].plot(kind="pie", title="Most MoH by Rank")
```

```python
# Plot 3
moh_df['awarded.date.year'].value_counts().plot(kind="area", title="Most MoH by State")
# This area plot of years awarded shows how badl the "63" screws up the data.  If I were to use this data in the future, I'd manually go in and correct the line and re-import the file.  I did not here becuase it created an interesting problem.
```
